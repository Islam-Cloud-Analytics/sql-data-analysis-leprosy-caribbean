```sql
SELECT
  c.country_name,
  COUNT(*) AS total_treatments,
  SUM(
    CASE
      WHEN t.outcome IN ('Cured', 'Completed') THEN 1
      ELSE 0
    END
  ) AS successful_treatments,
  ROUND(
    SUM(
      CASE
        WHEN t.outcome IN ('Cured', 'Completed') THEN 1
        ELSE 0
      END
    )
    * 100.0
    / COUNT(*),
    1
  ) AS success_rate_percent
FROM
  `data-analysis-project-catalog`.`sql-data-analysis-leprosy-caribbean`.`fact_treatment_episode` t
    JOIN `data-analysis-project-catalog`.`sql-data-analysis-leprosy-caribbean`.`fact_case_registration` r
      ON t.registration_id = r.registration_id
    JOIN `data-analysis-project-catalog`.`sql-data-analysis-leprosy-caribbean`.`dim_health_facility` h
      ON r.facility_id = h.facility_id
    JOIN `data-analysis-project-catalog`.`sql-data-analysis-leprosy-caribbean`.`dim_subnational_unit` s
      ON h.subnational_id = s.subnational_id
    JOIN `data-analysis-project-catalog`.`sql-data-analysis-leprosy-caribbean`.`dim_country` c
      ON s.country_id = c.country_id
GROUP BY
  c.country_name
ORDER BY
  success_rate_percent DESC
```
