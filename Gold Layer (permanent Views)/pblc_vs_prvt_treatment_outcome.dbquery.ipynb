```sql
SELECT
  h.is_public,
  r.who_disability_grade,
  COUNT(*) AS total_treatments,
  SUM(
    CASE
      WHEN t.outcome IN ('Cured', 'Completed') THEN 1
      ELSE 0
    END
  ) AS successful_treatment,
  ROUND(
    SUM(
      CASE
        WHEN t.outcome IN ('Cured', 'Completed') THEN 1
        ELSE 0
      END
    )
    / 100
    * COUNT(*),
    1
  ) AS success_rate_percent
FROM
  `data-analysis-project-catalog`.`sql-data-analysis-leprosy-caribbean`.fact_case_registration r
    JOIN `data-analysis-project-catalog`.`sql-data-analysis-leprosy-caribbean`.dim_health_facility h
      ON r.facility_id = h.facility_id
    JOIN `data-analysis-project-catalog`.`sql-data-analysis-leprosy-caribbean`.fact_treatment_episode t
      ON t.registration_id = r.registration_id
GROUP BY
  h.is_public,
  r.who_disability_grade
ORDER BY
  h.is_public,
  r.who_disability_grade
```
