```sql
SELECT
  detected_via AS detection_method,
  registration_type,
  SUM(
    CASE
      WHEN who_disability_grade = 0 THEN 1
      ELSE 0
    END
  ) AS grade_0,
  SUM(
    CASE
      WHEN who_disability_grade = 1 THEN 1
      ELSE 0
    END
  ) AS grade_1,
  SUM(
    CASE
      WHEN who_disability_grade = 2 THEN 1
      ELSE 0
    END
  ) AS grade_2
FROM
  `data-analysis-project-catalog`.`sql-data-analysis-leprosy-caribbean`.fact_case_registration
GROUP BY
  detected_via,
  registration_type
ORDER BY
  detected_via,
  registration_type
```
